- Severity of Illness vs Anxiety Level: Positive Relationship. As people become sicker, anxiety levels increase.


- Satisfaction vs Patient's Age: Negative Relationship. As people get older, their satisfaction level decreases.

```r
fit <- lm(V1~V2+V3+V4, data=satisfaction)
```

V1 is the Satisfaction

V2 is the Patient's Age

V3 is the Severity of Illness

V4 is the Anxiety Level

```r
summary(fit)
```

Estimated linear regression line is:

Satisfaction = 158.4913 - 1.1416 * Patient's Age - 0.4420 * Severity of Illness - 13.4702 * Anxiety Level

```r
residuals(fit)
```

```r
boxplot(residuals(fit))
```

There don't seem to be any outliers

```r
resid = residuals(fit)
```

```r
plot(age, resid)
```

```r
plot(severity, resid)
```

```r
plot(anxiety, resid)
```

```r
qqnorm(resid)
```

- There seems to be random trend between residuals vs age, residuals vs severity, and residuals vs anxiety.

- Normal QQplot suggests that residuals are normal due to linear behaviour
